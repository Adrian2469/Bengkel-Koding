```python
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower) & (df[col] <= upper)]
```

# Pemodelan

```python
X = df.drop(columns=['obesity'])  # Masukkan semua fitur kecuali target
y = df['obesity']
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_smote)
X_test_scaled = scaler.transform(X_test)
```

Memasukkan kolom input ke X dan kolom target ke y

Menggunakan smote untuk menyeimbangkan data supaya hasilnya tidak berat sebelah ke data mayoritas

Normalisasi data menggunakan StandardScaler untuk menyetarakan skala semua fitur

```python
def evaluate_model(model, X_train_scaled, X_test_scaled, y_train_smote, y_test):
    start = time.perf_counter()
    model.fit(X_train_scaled, y_train_smote)   # gunakan X_train_scaled yang benar
    end = time.perf_counter()
    train_time = end - start

    y_pred = model.predict(X_test_scaled)      # gunakan X_test_scaled yang benar

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
    recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
    f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)
    report = classification_report(y_test, y_pred, zero_division=0)

    return accuracy, precision, recall, f1, report, train_time
```

Pemilihan model

```python
models = {
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "KNN" : KNeighborsClassifier(),
    "Logistic Regression": LogisticRegression(),
    "SVM": SVC()
}
```

```python
result = {}
for name, model in models.items():
    accuracy, precision, recall, f1, report, train_time = evaluate_model(model, X_train_scaled, X_test_scaled, y_train_smote, y_test)

    # Gabungkan report dan training time
    extended_report = report + f"\nTraining Time: {train_time:.9f} seconds"

    result[name] = (accuracy, precision, recall, f1, extended_report)
```

# Evaluasi Model

```python
for name,(accuracy,precision,recall,f1,report) in result.items():
  print(f"Model: {name}")
  print("Classification Report:")
  print(report)
  print("-" *50)
```

Dari hasil diatas kita bisa lihat performa setiap model

```python
plt.figure(figsize=(18, 12))

for i, (name, model) in enumerate(models.items(), 1):
    start = time.time()
    model.fit(X_train_scaled, y_train_smote)
    end = time.time()

    y_pred = model.predict(X_test_scaled)

    acc = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred, output_dict=True)
    cm = confusion_matrix(y_test, y_pred)

    result[name] = {
        'accuracy': acc,
        'precision': report['weighted avg']['precision'],
        'recall': report['weighted avg']['recall'],
        'f1-score': report['weighted avg']['f1-score'],
        'train_time': end - start
    }

    # Gambar confusion matrix
    plt.subplot(3, 2, i)
    sns.heatmap(cm, annot=True, fmt='d', cmap='coolwarm')
    plt.title(f'Confusion Matrix - {name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')

plt.tight_layout()
plt.show()
```

Dari Confusion Matrix diatas kita bisa lihat performa prediksi dari suatu model

# Hyperparameter Tuning

Disini saya tuning SVM dan RandomForest karena memiliki f1 score yang menjanjikan.


Model yang mengalami peningkatan setelah di tuning akan menjadi model yang saya gunakan untuk deployment

```python
param_grid_svm = {
    'C': [0.1, 1, 10],
    'gamma': ['scale', 0.001, 0.01, 0.1, 1],
    'kernel': ['rbf', 'linear']
}

grid_svm = GridSearchCV(SVC(), param_grid_svm, cv=5, scoring='f1_weighted', n_jobs=-1, verbose=1)
grid_svm.fit(X_train_scaled, y_train_smote)
print("Best SVM Params:", grid_svm.best_params_)
```

```python
param_grid_rf = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'bootstrap': [True, False]
}

grid_rf = GridSearchCV(RandomForestClassifier(), param_grid_rf, cv=5, scoring='f1_weighted', n_jobs=-1, verbose=1)
grid_rf.fit(X_train_scaled, y_train_smote)
print("Best RF Params:", grid_rf.best_params_)
```

Menggunakan GridSearch untuk menemukan parameter apa saja yang cocok untuk mendorong performa model

```python
# SVM
best_svm = grid_svm.best_estimator_
y_pred_svm = best_svm.predict(X_test_scaled)

# RF
best_rf = grid_rf.best_estimator_
y_pred_rf = best_rf.predict(X_test_scaled)
```

```python
# SVM
print("SVM Best Model")
print("Accuracy:", accuracy_score(y_test, y_pred_svm))
print("Classification Report:\n", classification_report(y_test, y_pred_svm))
sns.heatmap(confusion_matrix(y_test, y_pred_svm), annot=True, fmt='d')
plt.title("Confusion Matrix - Best SVM")
plt.show()

# RF
print("Random Forest Best Model")
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("Classification Report:\n", classification_report(y_test, y_pred_rf))
sns.heatmap(confusion_matrix(y_test, y_pred_rf), annot=True, fmt='d')
plt.title("Confusion Matrix - Best Random Forest")
plt.show()
```

SVM mengalami peningkatan yang cukup signifikan sementara RF tidak, sehingga SVM akan menjadi model yang akan saya gunakan
